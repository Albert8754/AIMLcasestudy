```python
# as per requirements Lets Divide the data to price_range column as label and remaining columns as data.
X = df.iloc[:,0:20]
y = df.iloc[:,-1]
```

```python
# Lets do some Feature Selection for better results and accuracy
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
kbest = SelectKBest(chi2,k=10)
best_feaures = kbest.fit(X,y)
best_feaures.scores_
```

```python
df_features = pd.DataFrame(best_feaures.scores_)
df_columns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([df_columns,df_features],axis=1)
featureScores.columns = ['Features','Score']
featureScores.sort_values(by='Score',ascending=False)
```

```python
X = df[['ram','px_height','battery_power','px_width','mobile_wt','int_memory','sc_w','talk_time','front_cam','sc_h']]
```

```python
# lets do some normalisation and scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Neural network Model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = Sequential()
model.add(Dense(6,activation='relu',input_dim = 10))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=100,validation_data=(X_test,y_test),batch_size=64)
```

```python
y_pred = model.predict(X_test)
```

Got an accuracy 0f 96.13

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Features','Price_Weight'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Features','Price_Weight'],loc='upper left')
plt.show()
```
